```python
y_train = train['SalePrice']
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(max_depth=15, n_estimators=100, random_state=42)
rf_regressor.fit(XX_train, y_train)
```

```python
train = pd.read_csv('C://Users//lxw//Desktop//STA//142A//project//house-prices-advanced-regression-techniques//train.csv')
test = pd.read_csv('C://Users//lxw//Desktop//STA//142A//project//house-prices-advanced-regression-techniques//test.csv')
```

```python
total = train.isna().sum().sort_values(ascending=False)
missing_data = pd.concat([total], axis=1, keys=["Total"])
## drop all columns with missing value ** this method should be considered later
train = train.drop((missing_data[missing_data.get("Total") > 1]).index, 1)
train = train.drop(train.loc[train.get("Electrical").isna()].index)
```

```python
## drop price and id. Price is the response variable 
X = train.drop(['Id', 'SalePrice'], axis=1)
```

```python
print(X.shape)
test = test.drop((missing_data[missing_data.get("Total") > 1]).index, 1)
print(test.shape)
```

```python
categories = list(train.select_dtypes(["object"]))

# Applying one hot encoding 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), categories)], remainder='passthrough')
```

```python
X = train.drop(['Id', 'SalePrice'], axis=1)
```

```python
X_test = test.drop(["Id"], axis=1)
```

```python
# Looping through all the missing data columns
for i in X_test.isna().columns:
    # Checking if the datatype is not an object and replacing it with mean value
    if X_test.dtypes[i] != "object":
        X_test[i] = X_test[i].fillna(X_test[i].mean())
    else:
        X_test[i] = X_test[i].fillna(X_test[i].mode()[0])
X_test.shape
```

```python
XX_train = ct.fit_transform(X)
XX_train
```

```python
XX_test = ct.transform(X_test)
```

```python
y = train.SalePrice
```

```python
XX_train
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(max_depth=15, n_estimators=100, random_state=42)
rf_regressor.fit(XX_train, y)
```

```python
y_preds = rf_regressor.predict(XX_test)
pd.set_option('display.max_rows',None)
```
